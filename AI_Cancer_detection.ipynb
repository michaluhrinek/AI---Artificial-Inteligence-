# AI - Cancer Detection

```python
import pandas as pd 
```

```python
dataset = pd.read_csv('cancer.csv')
```

```python
dataset
```

```python
x = dataset.drop(columns=["diagnosis(1=m, 0=b)"])  #it takes everything into x besides diagnosis
y = dataset["diagnosis(1=m, 0=b)"]   #it takes diagnosis into y
```

# Splitting data into testing and training set 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
!pip install tensorflow
```

# Build train model 

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(256, input_shape=(x_train.shape[1],), activation='sigmoid'))   
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))   #adding layers to our model
```

The line of code you provided is adding a dense (fully connected) layer to a Keras Sequential model.

Here's what each argument in the Dense layer does:

    256: The first argument is the number of neurons in the layer, in this case 256. The number of neurons in a layer determines the capacity of the model to learn more complex relationships between the inputs and outputs.

    input_shape: The second argument is the shape of the input data. In this case, x_train.shape is being used as the input shape. This means that the first layer of the model will have an input shape that matches the shape of the training data.

    activation: The activation function to be applied to the output of each neuron in the layer. In this case, the 'sigmoid' activation function is being used. The sigmoid function is often used in binary classification problems to squash the output values between 0 and 1.

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1000)  #epochs - how many times algorithm is going over and over data
```

# Evaluate the model 

```python
model.evaluate(x_test, y_test)
```

We can say that based on 92% accuracy AI can diagnose patient cancer or not

# Go over Entire dataset to see who has a cancer and who doesnt?

```python
# count the number of cancer and no-cancer instances
import pandas as pd

dataset = pd.read_csv('cancer.csv')

```

```python
dataset
```

```python
print(dataset.columns)
```

```python
num_cancer = 0
num_no_cancer = 0
for status in dataset['diagnosis(1=m, 0=b)']:
    if status == 1:
        num_cancer += 1
    else:
        num_no_cancer += 1

print(f"Number of people with cancer: {num_cancer}")
print(f"Number of people without cancer: {num_no_cancer}")
```
